```python
import matplotlib.pyplot as plt

# Dati delle altezze
altezze = [160, 163, 165, 167, 168, 168, 169, 169, 170, 170, 171, 172, 172, 174, 175, 176, 177, 178, 180, 182]

# Conta le frequenze delle altezze
frequenze = {}
for altezza in altezze:
    frequenze[altezza] = frequenze.get(altezza, 0) + 1

# Estrai altezze e relative frequenze
altezze_univoche = list(frequenze.keys())
frequenze_valori = list(frequenze.values())

# Plot del grafico a barre
plt.plot(altezze_univoche, frequenze_valori, color='skyblue')

# Imposta i titoli e le etichette
plt.title('Distribuzione delle Altezze')
plt.xlabel('Altezza (cm)')
plt.ylabel('Frequenza')

# Mostra il grafico
plt.show()
```

```python
import numpy as np

# Dati delle altezze
altezze = [160, 163, 165, 167, 168, 168, 169, 169, 170, 170, 171, 172, 172, 174, 175, 176, 177, 178, 180, 182]

# Calcola i quantili al 25°, 50° e 75° percentile
quantili = np.percentile(altezze, [25, 50, 75])

# Stampa i quantili calcolati
print("25° percentile (Primo quartile):", quantili[0])
print("50° percentile (Mediana):", quantili[1])
print("75° percentile (Terzo quartile):", quantili[2])
```

```python
import random

# Genera le età dei dipendenti in un intervallo plausibile per un adulto
eta_dipendenti = [random.randint(20, 65) for _ in range(30)]

# Calcola la media delle età dei dipendenti
media = sum(eta_dipendenti) / len(eta_dipendenti)

# Calcola la moda delle età dei dipendenti
moda = max(set(eta_dipendenti), key=eta_dipendenti.count)

# Calcola la mediana delle età dei dipendenti
eta_dipendenti_ordinato = sorted(eta_dipendenti)
lunghezza = len(eta_dipendenti_ordinato)
if lunghezza % 2 == 0:
    mediana = (eta_dipendenti_ordinato[lunghezza//2 - 1] + eta_dipendenti_ordinato[lunghezza//2]) / 2
else:
    mediana = eta_dipendenti_ordinato[lunghezza//2]

# Stampa la lista delle età dei dipendenti
print("Età dei dipendenti:", eta_dipendenti)

# Stampa la media, moda e mediana
print("Media delle età:", media)
print("Moda delle età:", moda)
print("Mediana delle età:", mediana)
```
