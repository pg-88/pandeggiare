Per rinominarle uso il metodo rename creando un dict da passare all'argomento columns del metodo.
Quindi:

1) recupero il nome attuale delle colonne 
2) recupe i nomi corretti delle colonne 
3) creo un dictionary 
4) uso il metodo rename
   

```python
# colonne attuali 
col = df.columns
nuovi_nomi = df.iloc[0]
d = {old: new for old, new in zip(col, nuovi_nomi)}
df.rename(columns=d, inplace=True)
df
```
